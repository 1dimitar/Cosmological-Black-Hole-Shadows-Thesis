```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

plt.rcParams["font.size"] = 18
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams["font.weight"] = "normal"
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['mathtext.fontset'] = 'stix'

def covert_to_KS(surface, theta, a):
    return [surface * np.cos(theta), np.sqrt(surface * surface + a * a * np.cos(theta-theta)) * np.sin(theta)]

def photon_orbit_polynomial(r, T, a, m):
    rho = r**2 + a**2 * np.cos(T)**2
    delta = r**2 + a**2 -2 * m * r
    return 4*r**2 *delta* a**2 * np.sin(T)**2 - (rho * (r - m) - 2 *r *delta)**2 



def plot_BH_regions(m, a, axis):
    theta = np.linspace(0, 2*np.pi, 1000)
    inner_region = []
    outer_region = []
    for T in theta:
        radii = fsolve(photon_orbit_polynomial, [2.4*m, 4*m], args=(T, a, m))
        inner_region.append(radii[0])
        outer_region.append(radii[1])
        
    outer_eh = (m + np.sqrt(m**2 - a**2))*np.cos(theta-theta)
    inner_eh = (m - np.sqrt(m**2 - a**2))*np.cos(theta-theta)
    ergosphere = m + np.sqrt(m**2 - a**2*np.cos(theta)**2)
   
    axis.plot(theta, ergosphere, 'blue', linewidth=4)
    axis.plot(theta, outer_eh, 'red', linewidth=4)
    axis.plot(theta, inner_eh, 'orange', linewidth=4)
    axis.plot(theta, inner_region, 'magenta', linewidth=4)
    axis.plot(theta, outer_region, 'magenta', linewidth=4) 
    
    axis.fill_between(theta, ergosphere, outer_eh, label=r'Ergosphere', color='cornflowerblue')
    axis.fill_between(theta, outer_eh, inner_eh, color='salmon', label=r'Outer Event Horizon, $r_+$')
    axis.fill(theta, inner_eh, color='gold', label=r'Inner Event Horizon, $r_-$')
    axis.fill_between(theta, inner_region, outer_region, color='magenta', label='Photon Region', alpha=0.4)

def kerr_surfaces(m, a):
    theta = np.linspace(0, 2*np.pi, 500)
    inner_region = []
    outer_region = []
    for T in theta:
        radii = fsolve(photon_orbit_polynomial, [2.4*m, 4*m], args=(T, a, m))
        inner_region.append(radii[0])
        outer_region.append(radii[1])

    outer_eh = (m + np.sqrt(m**2 - a**2))*np.cos(theta-theta)
    inner_eh = (m - np.sqrt(m**2 - a**2))*np.cos(theta-theta)
    ergosphere = m + np.sqrt(m**2 - a**2*np.cos(theta)**2)
    return outer_eh, inner_eh, ergosphere, inner_region, outer_region
    
def plot_kerr_schild(m, a):
    fig, axis = plt.subplots(nrows = 1, ncols=len(a), figsize=[2.5 * 7, 0.92 * 7])
    for (axis, a) in zip(axis, a):
        axis.set_xlim(-4.3*m, 4.3*m)
        axis.set_ylim(-4.3*m, 4.3*m)
        axis.axvline(0, color='gray')
        axis.axhline(0, color='gray')
        axis.set_xlabel('$x$ / [m]')
        axis.set_ylabel('$z$ / [m]')
        axis.set_title(fr"$a$ = {a}, $\quad m$ = {m}")
        
        theta = np.linspace(0, 2*np.pi, 500)
        outer_eh, inner_eh, ergosphere, inner_region, outer_region = kerr_surfaces(m, a)
        
        ergosphere_z, ergosphere_r = covert_to_KS(ergosphere, theta, a)
        outer_z, outer_r = covert_to_KS(outer_eh, theta, a)
        inner_z, inner_r = covert_to_KS(inner_eh, theta, a)
        ipr_z, ipr_r = covert_to_KS(np.array(inner_region), theta, a)
        singularity = (-a, a)
        
        opr_z, opr_r = covert_to_KS(np.array(outer_region), theta, a)
        axis.plot(ergosphere_r, ergosphere_z, 'blue', label=r'Ergosphere', linewidth=4)
        axis.plot(outer_r, outer_z, 'red', label=r'Outer Event Horizon, $r_+$', linewidth=4)
        axis.plot(inner_r, inner_z, 'orange', label=r'Inner Event Horizon, $r_-$', linewidth=4)
        axis.plot(ipr_r, ipr_z, 'magenta', linewidth=4)
        axis.scatter(singularity, (0, 0), c='k', linewidth=1, zorder=10, label="Curvature Singularity")
        axis.plot(opr_r, opr_z, 'magenta', linewidth=4)

        axis.fill(opr_r, opr_z, color='magenta', alpha=0.4)
        axis.fill(ipr_r, ipr_z, color='white')
        axis.fill(ergosphere_r, ergosphere_z, color='cornflowerblue', alpha=0.6)
        axis.fill(outer_r, outer_z, color='salmon')
        axis.fill(inner_r, inner_z, color='khaki')

    fig.tight_layout()
    fig.subplots_adjust(bottom=0.21)
    fig.legend(['Ergoregion', r'Outer Event Horizon, $r_{\text{eh}, +}$', r'Inner Event Horizon, $r_-$', 
                'Photon Region', r'Curvature Singularity'], ncols=5, loc=8, fontsize=20)
    
    # plt.savefig(f"Kerr Schild Surfaces.png", bbox_inches='tight', dpi=800)
    
def plot_boyer_lindquist(m, a):
    fig, ax = plt.subplots(nrows = 1, ncols=len(a), figsize=[21,6], subplot_kw={'projection': 'polar'}) #22
    for (axis, a) in zip(ax, a):
        axis.set_rmax(4.3)
        axis.grid(True)
        axis.set_rticks([0, 1, 2, 3, 4])  # Less radial ticks
        axis.set_theta_zero_location('N')
        axis.set_theta_direction(-1)
        axis.set_rlabel_position(22.5)  # Move radial labels away from plotted line
        axis.set_title(fr"$a$ = {a}, $\quad m$ = {m}")
        
        theta = np.linspace(0, 2*np.pi, 500)
        outer_eh, inner_eh, ergosphere, inner_region, outer_region = kerr_surfaces(m, a)
        axis.plot(theta, ergosphere, 'blue', linewidth=4)
        axis.plot(theta, outer_eh, 'red', linewidth=4)
        axis.plot(theta, inner_eh, 'orange', linewidth=4)
        axis.plot(theta, inner_region, 'magenta', linewidth=4)
        axis.plot(theta, outer_region, 'magenta', linewidth=4) 

        axis.fill_between(theta, ergosphere, outer_eh, label=r'Ergosphere', color='cornflowerblue')
        axis.fill_between(theta, outer_eh, inner_eh, color='salmon', label=r'Outer Event Horizon, $r_+$')
        axis.fill(theta, inner_eh, color='gold', label=r'Inner Event Horizon, $r_-$')
        axis.fill_between(theta, inner_region, outer_region, color='magenta', label='Photon Region', alpha=0.4)

    fig.legend([r'Ergoregion', r'Outer Event Horizon, $r_{\text{eh}, +}$',
                r'Inner Event Horizon, $r_{\text{eh}, -}$','Photon Region'], loc=8, ncols=4)
    
    # plt.savefig(f"Kerr Photon Region.png", bbox_inches='tight', dpi=400)
    
plot_boyer_lindquist(m=1, a=[0, 0.5, 0.95])
plot_kerr_schild(m=1, a=[0, 0.5, 0.95])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

np.seterr(divide = 'ignore') 
plt.rcParams["font.size"] = 22
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams["font.weight"] = "normal"
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['mathtext.fontset'] = 'stix'

def eqn_for_rp(rp, a, r0, m, T0, b):
    numerator = -(rp**3 - 3*rp**2*m + rp*a**2 + a**2 *m +  a**2*np.sin(T0)**2* (rp - m))
    denominator = (2*a*rp*np.sin(T0)*np.sqrt(rp**2 + a**2 - 2*m*rp))
    return numerator / denominator + b

def obtain_shadow(a, r0, m, T0):
    r_ph_min = fsolve(eqn_for_rp, [m, 3*m], args=(a, r0, m, T0, -1))[0]
    r_ph_max = fsolve(eqn_for_rp, [3*m, 4*m], args=(a, r0, m, T0, 1))[0]

    r_ph = np.linspace(r_ph_min, r_ph_max, int(1e6))
    SinGamma = 2*r_ph*np.sqrt(r_ph**2 + a**2 - 2*m*r_ph) * np.sqrt(r0**2 + a**2 - 2*m*r0) / (r0**2 *r_ph - r0**2 *m + r_ph**3 - 3*r_ph**2 *m + 2*r_ph *a**2)
    SinPsi = -(r_ph**3 - 3*r_ph**2*m + r_ph*a**2 + a**2 *m +  a**2*np.sin(T0)**2* (r_ph - m)) / (2*a*r_ph*np.sin(T0)*np.sqrt(r_ph**2 + a**2 - 2*m*r_ph))

    Gamma = np.arcsin(SinGamma)
    Psi = np.arcsin(SinPsi)

    x = -2 * np.tan(Gamma) * np.sin(Psi)
    y = -2 * np.tan(Gamma) * np.cos(Psi)
    return x, y

colors = ['black', 'darkred', 'red', 'darkorange', 'gold', 'lawngreen']
m_list = [1, 2, 5, 10, 20]
a_list = [0.01, 0.3, 0.6, 0.9, 0.999]
r0_list = [5, 10, 20, 50, 100]
T0_list = [0.01, np.pi/8, 2*np.pi/8, 3*np.pi/8, np.pi/2]

fig, axis = plt.subplots(nrows = 2, ncols=2, figsize=[20,20])

for i, m in enumerate(m_list):
    x, y = obtain_shadow(a=0.9999*m, r0=100, m=m, T0=np.pi/2)
    axis[0, 0].plot(x, y, label=fr'$m$ = {m}', color=colors[i], linewidth=2)
    axis[0, 0].plot(x, -y, color=colors[i], linewidth=2)
    
    axis[0, 0].set_title(r"$a = 0.999m, \ r_\text{O}=100,\  \theta_\text{O}=\frac{\pi}{2}$")
    axis[0, 0].set_xlabel(r"$x(r_\text{ps})$")
    axis[0, 0].set_ylabel(r"$y(r_\text{ps})$")
    axis[0, 0].axvline(0, color='k')
    axis[0, 0].axhline(0, color='k')
            
    axis_limit=8
    axis[0, 0].set_xlim(-0.15*axis_limit, axis_limit)
    axis[0, 0].set_ylim(-0.625*axis_limit, 0.625*axis_limit)
    axis[0, 0].legend(loc=3, fontsize=16, ncols=1)
    

for i, a in enumerate(a_list):
    x, y = obtain_shadow(a=a, r0=10, m=1, T0=np.pi/2)
    axis[0, 1].plot(x, y, label=fr'$a$ = {a}', color=colors[i], linewidth=2)
    axis[0, 1].plot(x, -y, color=colors[i], linewidth=2)
    
    axis[0, 1].set_title(r"$m = 1, \ r_\text{O}=10, \ \theta_\text{O}=\frac{\pi}{2}$")
    axis[0, 1].set_xlabel(r"$x(r_\text{ps})$")
    axis[0, 1].set_ylabel(r"$y(r_\text{ps})$")
    axis[0, 1].axvline(0, color='k')
    axis[0, 1].axhline(0, color='k')
            
    axis_limit=2
    axis[0, 1].set_xlim(-axis_limit, axis_limit)
    axis[0, 1].set_ylim(-axis_limit, axis_limit)
    axis[0, 1].legend(loc=3, fontsize=16, ncols=1)

    
for i, r0 in enumerate(r0_list):
    x, y = obtain_shadow(a=0.999, r0=r0, m=1, T0=np.pi/2)
    axis[1, 0].plot(x, y, label=r'$r_\text{O}$ = '+ str(r0), color=colors[i], linewidth=2)
    axis[1, 0].plot(x, -y, color=colors[i], linewidth=2)
    
    axis[1, 0].set_title(r"$m = 1, \ a = 0.999, \ \theta_\text{O}=\frac{\pi}{2}$")
    axis[1, 0].set_xlabel(r"$x(r_\text{ps})$")
    axis[1, 0].set_ylabel(r"$y(r_\text{ps})$")
    axis[1, 0].axvline(0, color='k')
    axis[1, 0].axhline(0, color='k')
    
    axis_limit=9
    axis[1, 0].set_xlim(-0.1*axis_limit, 0.9*axis_limit)
    axis[1, 0].set_ylim(-0.5*axis_limit, 0.5*axis_limit)
    axis[1, 0].legend(loc=3, fontsize=16, ncols=1)
    
    
for i, T0 in enumerate(T0_list):
    x, y = obtain_shadow(a=0.999, r0=10, m=1, T0=T0)
    axis[1, 1].plot(x, y, label=r'$\theta_\text{O}$ = '+ str(round(T0, 3)), color=colors[i], linewidth=2)
    axis[1, 1].plot(x, -y, color=colors[i], linewidth=2)
    
    axis[1, 1].set_title(r"$m = 1, \ a = 0.999, \ r_\text{O}=10$")
    axis[1, 1].set_xlabel(r"$x(r_\text{ps})$")
    axis[1, 1].set_ylabel(r"$y(r_\text{ps})$")
    axis[1, 1].axvline(0, color='k')
    axis[1, 1].axhline(0, color='k')
    
    axis_limit=2
    axis[1, 1].set_xlim(-0.75*axis_limit, axis_limit)
    axis[1, 1].set_ylim(-0.875*axis_limit, 0.875*axis_limit)
    axis[1, 1].legend(loc=3, fontsize=16, ncols=1)
    
# plt.savefig(f"Kerr BH Shadow.png", bbox_inches='tight', dpi=400)
```

# Plot 29

```python
import matplotlib.pyplot as plt
import math
import numpy as np

plt.rcParams["font.size"] = 20
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams["font.weight"] = "normal"
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['mathtext.fontset'] = 'stix'

def static_kottler(m, H0, r0):
    num = (1 - 2 * m / r0 - H0**2 * r0**2)
    den = (1 - 27 * H0 ** 2 * m**2)
    kottler_shadow = np.arcsin( (3 * m * np.sqrt(3) / r0) * np.sqrt(num / den) )
    return kottler_shadow

def comoving_kottler(m, H0, r0):
    term1 = np.sqrt(27) * m / r0
    term2 = np.sqrt(1 - 2 * m / r0)
    term3 = np.sqrt(1 - 27 * H0 **2 * m **2)
    term4 = np.sqrt(27) * m * H0
    term5 = np.sqrt(1 - 27 * m **2 * (1 - 2 * m / r0) / r0 ** 2)
    kottler_shadow = np.arcsin(term1 * term2 * term3 + term4 * term5)
    return kottler_shadow

def plot_Kottler_shadow(m, L, axis, color, r_lim):
    H0 = np.sqrt(L / 3)
    a = (1 / 3) * np.arccos(np.sqrt(27) * m * H0)
    inner_eh = (2 / (np.sqrt(3) * H0)) * np.cos(a + np.pi / 3)
    outer_eh = (2 / (np.sqrt(3) * H0)) * np.cos(a - np.pi / 3)
    r0 = np.linspace(inner_eh, r_lim, 100000)
    axis[0].plot(r0, static_kottler(m=m, H0=H0, r0=r0), color=color, label=fr"$\Lambda$ = {L}", linewidth=3)
    axis[1].plot(r0, comoving_kottler(m=m, H0=H0, r0=r0), label=fr"$\Lambda$ = {L}", color=color, linewidth=3)

    
def plot_figure(m, lambda_list, upper_bound):
    Lambda_list = [L for L in lambda_list if L < (1 / (9 * m * m))]
    if Lambda_list != lambda_list: print(f'Parameters L = {set(Lambda_list) ^ set(lambda_list)} exceed 1 / 9m^2')
    
    color_list = ['brown', 'red', 'darkorange', 'gold', 'limegreen']
    r0 = np.linspace(3 * m, 100 * m, 10000)[1:-1]
    photon_sphere = 3 * m
    schw_shadow = np.arcsin( (3 * np.sqrt(3) * m / r0) * np.sqrt(1 - 2 * m / r0) )
    
    fig, axis = plt.subplots(nrows = 2, ncols=1, figsize=(13, 13), facecolor='white')

    for ax in axis:
        ax.set_xlim(0.94*photon_sphere, upper_bound)
        ax.set_ylim(0, 1.07*max(schw_shadow))
        ax.set_xlabel(r"Observer distance, $r_\text{O}$ \ [m]")
        ax.axvline(3*m, color='k', linestyle='--', label=r'Photon Sphere, $r_{\text{ps}}$', linewidth=3)

    axis[0].set_title('Static Observer')
    axis[0].set_ylabel(r"Angular size, $\alpha_\text{sh, static}$")
    for i, L in enumerate(Lambda_list):  
        plot_Kottler_shadow(m=m, L=L, axis=axis, color=color_list[i], r_lim=upper_bound)
    axis[0].plot(r0, schw_shadow, 'b--', label="Schwarzschild", linewidth=4)
    axis[1].plot(r0, schw_shadow, 'b--', label="Schwarzschild", linewidth=4)


    axis[1].set_title('Comoving Observer')
    axis[1].set_ylabel(r"Angular size, $\alpha_\text{sh, comov}$")
    
    fig.tight_layout()
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    axis[0].legend(*zip(*unique), loc=1)
    # plt.savefig(fr"C:\Users\dimit\Computational Physics\Thesis\Kottler Static and Comoving.png", bbox_inches='tight', dpi=500)

plot_figure(m=1, lambda_list=[1e-2, 1e-3, 5e-4, 1e-7], upper_bound = 80)
```
